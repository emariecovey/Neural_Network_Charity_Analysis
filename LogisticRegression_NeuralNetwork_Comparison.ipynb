```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
diabetes_df = pd.read_csv('Resources/diabetes.csv')
diabetes_df
```

```python
# No preprocessing or scaling required for the logistic regression model, but we do need to scale for neural network

X = diabetes_df.drop(columns=["Outcome"])
y = diabetes_df["Outcome"]

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Preprocess numerical data for neural network

X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating logistic regression model
model = LogisticRegression(solver="lbfgs",max_iter=200).fit(X_train, y_train)

predictions = model.predict(X_test)

print(f" Logistic regression model accuracy: {accuracy_score(y_test,predictions):.3f}")
```

```python
# Define the basic neural network model
# the single hidden layer has 8 variables going into it, and 16 neurons, and uses relu activation function
# output layer has one outcome (neuron?) and uses sigmoid activation because we're looking at a binary outcome (diabetes or no diabetes). Logistic regression also uses sigmoid function
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu", input_dim=8))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics (use binary_crossentropy and accuracy when predicting binary outcome)
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model (model only going through data 50 times not 200 like logistic model, so less chance of overfitting)
nn_model.fit(X_train_scaled, y_train, epochs=50)

# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

```
