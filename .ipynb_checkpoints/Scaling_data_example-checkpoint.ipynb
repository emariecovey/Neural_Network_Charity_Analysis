```python
# Import our dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Read in our dataset
hr_df = pd.read_csv("Resources/hr_dataset.csv")
hr_df
```

```python
# Make sure to do one-hot encoding if there are categorical variables first

# Create the StandardScaler instance
scaled_data = StandardScaler().fit_transform(hr_df)

# Make dataframe from scaled data
scaled_df = pd.DataFrame(scaled_data, columns = hr_df.columns)
scaled_df
```

```python
# Checking to make sure mean is about 0 and SD is 1
scaled_df.describe()
```
